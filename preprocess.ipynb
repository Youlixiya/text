```python
import jieba
import jieba.analyse
import pandas as pd
```

```python
#先将四个岗位的数据读进来然后拼在一起用于后面统一处理，词向量的训练使用四个岗位的数据一起
analyze = pd.read_table('datas/analyze.txt',names=['des'], header=None)
analyze['target'] = [0] * analyze.shape[0]
algorithm = pd.read_table('datas/algorithm.TXT',names=['des'],header=None)
algorithm['target'] = [1] * algorithm.shape[0]
develop = pd.read_table('datas/develop.TXT',names=['des'],header=None)
develop['target'] = [2] * develop.shape[0]
operate = pd.read_table('datas/operate.TXT',names=['des'],header=None)
operate['target'] = [3] * operate.shape[0]
df = pd.concat([analyze, algorithm, develop, operate], axis=0)
df = df.reset_index(drop=True)
df
```

```python
jieba.load_userdict('datas/phrase.txt')
```

```python
df["des"] = df["des"].map(lambda x: " ".join(jieba.lcut(x)))
df
```

```python
def load_stop_words():
    with open('datas/stop_words.txt',encoding='utf-8') as fr:
        stop_words = set([word.strip() for word in fr])
    return stop_words
# 加载停用词
stop_words = load_stop_words()
# 切词并过滤调停用词
df["des"] = df["des"].map(lambda x: "".join([i for i in jieba.cut(x) if i not in stop_words]))
# 保存处理好的文本
df.to_csv("datas/lcut.csv", index=False, header=False, columns=['des', 'target'])
```

```python
#统计所有语料中出现的词的数量
def count_words_num():
    words_num_dict = {}
    for i in range(df["des"].shape[0]):
        for word in df["des"][i].strip().split():
            words_num_dict[word] = words_num_dict.get(word, 0) + 1
    return words_num_dict
words_num_dict = count_words_num()
words_num_dict, len(words_num_dict)
```

```python
#获取word2index字典将每一个词转化为对应的编号
#获取index2word字典将编号转化为对应的词
word2index = {k : i for i, k in enumerate(words_num_dict.keys())}
index2word = {v : k for k, v in word2index.items()}
word2index
```

```python
len(word2index), len(index2word)
```

```python
#将每一行数据都转成对应的编号后输出到csv文件中
def func(x):
    res = []
    for i in x.split():
        res.append(word2index[i])
    return res
df["des"] = df["des"].apply(lambda x : func(x))
df.to_csv("datas/datas_index.csv", index=False, header=False, columns=['des', 'target'])
df
```
